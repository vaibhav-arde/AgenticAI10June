## While Loop

```python
while condition:
    code
```

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
while True:
    my_number = int(input("Enter some number : "))

    if my_number == 7:
        print(f"You guessed correct number that is {my_number}")
        break
    print(f"Your number is : {my_number}")
```

# Dictionaries


```python
student = {
    "name": "Rahul",
    "age": 20,
    "city": "Pune"
}
```

```python
print(student["name"])
```

```python
student["course"]= "Agentic AI"
print(student)
```

```python
for key, value in student.items():
    print(key, value)
```

```python
for key in student.keys():
    print(key)
```

```python
for i in student.values():
    print(i)
```

# List Comprehensions

```python
squares = []

for i in range(1, 6):
    squares.append(i * i)

print(squares)
```

```python
squares = [i*i for i in range(1, 6)]

print(squares)
```

```python
squaresEvens = [i*i for i in range(1, 11) if i % 2 == 0]

print(squaresEvens)
```
