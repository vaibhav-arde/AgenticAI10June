## Example 1: OpenAI API — Chat Completion

OpenAI is the industry standard for LLM APIs.
- `client`: Your connection to OpenAI's servers
- `model`: Which AI model to use (`gpt-4o-mini` is standard, fast, and cheap)
- `messages`: The conversation in role/content format
- `temperature`: Creativity control (0=focused, 1=creative)

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()  # Load API keys from .env file

# Step 1: Create client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# Step 2: Make a chat completion request
response = client.chat.completions.create(
    model="gpt-4o-mini",        # Fast and cost-efficient
    messages=[
        {
            "role": "system",     # Instructions for the AI
            "content": "You are a senior Python developer. Give concise, practical answers."
        },
        {
            "role": "user",       # The human's question
            "content": "What are the top 3 Python libraries for data science and why?"
        }
    ],
    temperature=0.7,              # Balanced creativity
    max_tokens=500                # Limit output length
)

# Step 3: Print response
print("🤖 OpenAI Response:")
print(response.choices[0].message.content)

# Step 4: Check token usage and costs
print(f"\n📊 Token Usage:")
print(f"  Input tokens:  {response.usage.prompt_tokens}")
print(f"  Output tokens: {response.usage.completion_tokens}")
print(f"  Total tokens:  {response.usage.total_tokens}")
print(f"  Estimated cost: ${(response.usage.prompt_tokens * 0.15 + response.usage.completion_tokens * 0.60) / 1_000_000:.6f} USD")
````

## Example 2: Google Gemini API

Google Gemini offers a massive 2 million token context window and a generous free tier for developers.

````python
import os
import google.generativeai as genai
from dotenv import load_dotenv

load_dotenv()

# Step 1: Configure with API key
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

# Step 2: Create model instance
model = genai.GenerativeModel("gemini-1.5-flash")

# Step 3: Generate content
response = model.generate_content(
    "Explain the difference between a list and a tuple in Python. "
    "Give a practical example for each."
)

# Step 4: Print response
print("🤖 Gemini Response:")
print(response.text)

# Step 5: Check token usage
print(f"\n📊 Token Usage:")
print(f"  Input tokens:  {response.usage_metadata.prompt_token_count}")
print(f"  Output tokens: {response.usage_metadata.candidates_token_count}")
````

## Example 3: Anthropic Claude API

Claude is known for state-of-the-art coding abilities, complex logical reasoning, and long-context capabilities.

````python
import os
import anthropic
from dotenv import load_dotenv

load_dotenv()

# Step 1: Create client
client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))

# Step 2: Create message
message = client.messages.create(
    model="claude-3-5-sonnet-latest",  # Latest Claude model
    max_tokens=1024,
    system="You are an expert Python code reviewer. Be concise and practical.",
    messages=[
        {
            "role": "user",
            "content": "Review this code and suggest improvements:\n\n"
                       "def calc(a,b,op):\n"
                       "    if op == '+':\n"
                       "        return a+b\n"
                       "    if op == '-':\n"
                       "        return a-b\n"
                       "    if op == '*':\n"
                       "        return a*b\n"
                       "    if op == '/':\n"
                       "        return a/b"
        }
    ]
)

# Step 3: Print response
print("🤖 Claude Response:")
print(message.content[0].text)

# Step 4: Token usage
print(f"\n📊 Token Usage:")
print(f"  Input tokens:  {message.usage.input_tokens}")
print(f"  Output tokens: {message.usage.output_tokens}")
````

## Example 4: Groq API — Lightning Fast Inference

## Example 6: Ollama — Run LLMs Locally

Runs models on your local hardware completely free, offline, and with 100% data privacy.

````python
import requests
import json

def chat_with_ollama(message: str, model: str = "gpt-oss:120b-cloud") -> str:
    """Send a message to locally running Ollama."""
    try:
        response = requests.post(
            "http://localhost:11434/api/chat",
            json={
                "model": model,
                "messages": [
                    {"role": "user", "content": message}
                ],
                "stream": False
            },
            timeout=60
        )
        response.raise_for_status()
        return response.json()["message"]["content"]
    except requests.ConnectionError:
        return "❌ Error: Ollama is not running. Start it with: ollama serve"
    except Exception as e:
        return f"❌ Error: {str(e)}"

print("🤖 Ollama Response (Local LLM):")
result = chat_with_ollama("What is a dictionary in Python? Give a quick example.")
print(result)
````

## Demo 1: Token Counting — See Tokens in Action

Understand BPE tokenization by checking how strings are broken down.

````python
import tiktoken

# Get the tokenizer for GPT-4o
encoder = tiktoken.encoding_for_model("gpt-4o")

texts = [
    "Hello World",
    "I am learning about LLMs",
    "supercalifragilisticexpialidocious",
    "def hello_world():\n    print('Hello!')",
    "मैं हिंदी में लिख रहा हूं"  # Hindi uses 2-3x more tokens!
]

print("=" * 60)
print(f"{'Text':<45} | {'Tokens':>6} | Token IDs")
print("=" * 60)

for text in texts:
    tokens = encoder.encode(text)
    print(f"{text:<45} | {len(tokens):>6} | {tokens[:5]}...")
print("=" * 60)
````

## Demo 2: Temperature Comparison

Observe how temperature settings affect the diversity and randomness of responses.

````python
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

prompt = "Write a one-sentence description of Python programming language."
temperatures = [0.0, 0.7, 1.5]

print("🌡️ TEMPERATURE COMPARISON\nPrompt:", prompt)
for temp in temperatures:
    print(f"\n🌡️ Temperature: {temp}")
    for i in range(2):
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}],
            temperature=temp,
            max_tokens=60
        )
        print(f"  Run {i+1}: {response.choices[0].message.content.strip()}")
````

## Demo 3: Hallucination Detection & Mitigation

Test how models respond to fake questions and how system instructions can reduce hallucinations.

````python
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

question = "Who wrote the book 'The Azure Chronicles of Digital Wisdom'?"  # Fake book!

system_prompt = """You are a helpful, honest assistant.
If you're not sure about something, say "I'm not certain about this." 
Never make up facts or citations.
"""

# 1. Without Guardrails
res1 = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": question}],
    temperature=0.7
)
print("⚠️ WITHOUT GUARDRAILS:\n", res1.choices[0].message.content.strip())

# 2. With Guardrails
res2 = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": question}
    ],
    temperature=0.0  # Zero temperature helps facts
)
print("\n✅ WITH GUARDRAILS & TEMP=0:\n", res2.choices[0].message.content.strip())
````

# 🤖 PART 3: Mini Project — AI Chatbot (CodeBuddy)

## Build an AI Chatbot with Memory

Implement conversation memory by maintaining message history.

````python
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

SYSTEM_PROMPT = """You are CodeBuddy, a friendly AI coding assistant.
- You explain things simply, like talking to a friend.
- You use emojis occasionally to be friendly 😊.
- You always encourage the learner.
"""

class AIChatbot:
    def __init__(self):
        self.history = [{"role": "system", "content": SYSTEM_PROMPT}]
        
    def send_message(self, message: str) -> str:
        self.history.append({"role": "user", "content": message})
        
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=self.history,
            temperature=0.7
        )
        reply = response.choices[0].message.content
        self.history.append({"role": "assistant", "content": reply})
        return reply

# Test conversation showing memory
bot = AIChatbot()
print("1. User: Hi, I am Alice!")
print("CodeBuddy:", bot.send_message("Hi, I am Alice!"))

print("\n2. User: What is my name?")
print("CodeBuddy:", bot.send_message("What is my name?"))
````

# 🎯 PART 4: Industry Project — AI Career Mentor

## AI Career Mentor

A production-style application containing system prompt templates, session analytics, and assessment mode.

````python
import os
import json
from datetime import datetime
from groq import Groq
from dotenv import load_dotenv

load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

CAREER_MENTOR_PROMPT = """You are CareerAI, a tech career mentor with 15+ years of experience.
You guide professionals into AI/ML, Full-Stack, and DevOps roles.
Provide specific, actionable steps. Remind them GitHub profile is crucial.
Always ask at least 2 questions about their background before advising.
"""

class CareerMentor:
    def __init__(self):
        self.conversation = [{"role": "system", "content": CAREER_MENTOR_PROMPT}]
        self.message_count = 0
        self.total_tokens = 0
        self.session_start = datetime.now()
        
    def chat(self, user_message: str) -> str:
        self.message_count += 1
        self.conversation.append({"role": "user", "content": user_message})
        
        try:
            response = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=self.conversation,
                temperature=0.7,
                max_tokens=1000
            )
            ai_response = response.choices[0].message.content
            self.conversation.append({"role": "assistant", "content": ai_response})
            self.total_tokens += response.usage.total_tokens
            return ai_response
        except Exception as e:
            self.conversation.pop()
            return f"Connection issue: {str(e)}"
            
    def get_analytics(self) -> str:
        duration = (datetime.now() - self.session_start).seconds
        cost = (self.total_tokens / 1_000_000) * 0.59  # LLaMA 70B pricing
        return (f"📊 SESSION STATS\n"
                f"   Duration: {duration}s | Messages: {self.message_count}\n"
                f"   Approx. Tokens: {self.total_tokens} | Estimated Cost: ${cost:.6f}")
````

````python
# Instantiating and Testing CareerMentor in Assessment Mode
mentor = CareerMentor()
print("🤖 Initial Greeting:")
print(mentor.chat("Hi, introduce yourself as CareerAI and ask me about my background."))
````
