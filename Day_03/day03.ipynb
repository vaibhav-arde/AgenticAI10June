# **Four Pillars of OOPS**

1. Encapsulation
2. Inheritance
3. Polymorphism
4. Abstraction

```python

class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance


account = BankAccount()
print(account.get_balance())
account.deposit(5000)
print(account.get_balance())
```

```python
class Animal:
    def eat(self):
        print("Eating")


class Dog(Animal):
    def bark(self):
        print("Barking")


dog = Dog()
dog.eat()
dog.bark()
```

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

    def test(self):
        print("Hello")

class Car(Vehicle):
    def start(self):
        print("Car Started")


car = Car()
car.start()
```

```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        print(self.name, self.salary)


emp1 = Employee("Rahul", 50000)
emp2 = Employee("Amit", 70000)

emp1.display()
emp2.display()
```

## Exception Handling

## **Common Exceptions**

| **Exception** | **Example** |
| ------------------- | ----------------- |
| ZeroDivisionError   | 10/0              |
| ValueError          | int(“abc”)      |
| IndexError          | list[10]          |
| KeyError            | dict[“x”]       |
| FileNotFoundError   | missing file      |

```python
try:
    num = int(input("Enter Number: "))
    print(100 / num)

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    num = int(input("Enter Number: "))
    print(100 / num)

except ValueError:
    print("Invalid Number")

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    num = int(input("Enter Number: "))
    result = 100 / num
    for i in range(5):
        if i == 3:
            break
        print(i)

except Exception as e:
    print(e)

else:
    print("Result:", result)

finally:
    print("Hi I am THE finally")
```

```python
try:
    file = open("data.txt")

except FileNotFoundError:
    print("File Not Found")

finally:
    print("Execution Completed")
```

# **File Modes**

| **Mode** | **Purpose** |
| -------------- | ----------------- |
| r              | Read              |
| w              | Write             |
| a              | Append            |
| x              | Create            |
| rb             | Read Binary       |
| wb             | Write Binary      |

```python
file = open("students.txt", "w")
file.write("Welcome to Python")
file.close()
```

```python
file = open("students.txt", "r")
content = file.read()
print(content)
file.close()
```

```python
file = open("students.txt", "a")
file.write("\nRahul")
file.close()
```

```python
with open("students.txt", "r") as file:
    content = file.read()

print(content)
```

```python
with open("students.txt", "w") as file:
    file.write("Siddhi")
```

```python
with open("students.txt", "r") as file:
    data = file.read()

print(data)
```

```python
students = ["Rahul", "Amit", "John"]

with open("students.txt", "w") as file:

    for student in students:
        file.write(student + "\n")
```
